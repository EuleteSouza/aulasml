## Treinamento do Modelo

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

## Prever um animal

```python
# Features [1 sim , 0 nao]
# tem bico?
# risco extino?
# faz cocoric?
# tem bico, esta em extino e no faz cocoric
#prever o papagaio

animal_misterioso = [1,1,0]
previsao = modelo.predict([animal_misterioso])
```

# Calculando a Acurcia do Modelo

## Dados e Rtulos

```python
# dados de treino x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
# rotulos (classes) de treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
# previsoes - dados de teste
teste_x = [misterio1, misterio2, misterio3]

#prevendo os testes
previsoes = modelo.predict(teste_x)
```

```python
# rotulos (classes) de teste (avaliao - aqui  a sua provinha)
teste_y = [0,1,1]
```

```python
# importando mdulo que j traz a funo de acurcia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma variavel para receber o retorno da funo acurcia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes)* 100

# mostrando a acurcia
print(f'Acurcia> {taxa_de_acerto:.2f}%')
```
